```python
import os

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import collections
import csv

from pprint import pprint as pp
import circlify as circ

matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

DIR = os.getcwd()
```

# 2020 contributions

```python
# get data 
df20 = pd.read_csv(DIR + "/2019to2020contributions.csv") 
df20

#extract unique professions
df20_nodups = df20.drop_duplicates(subset=["contributor_last_name", "contributor_middle_name",
                                              "contributor_middle_name"])
occs20 = collections.Counter(df20_nodups["contributor_occupation"])

with open('2020occs.csv','w') as csvfile:
    writer=csv.writer(csvfile)
    for key, value in occs20.items():
        writer.writerow([key] + [value]) 
```

there is a step in here where I took the 2020occs.csv file and manually sorted the professions into the alloccsSORTED spreadsheet

```python
# faculty, teaching positions, admin, academia & research, medical & health, undergrads, grad students, other
# numbers from alloccsSORTED spreadsheet

circs20 = sorted([740,250,1058,548,510,273,564,140])

circles20 = circ.circlify(circs20, show_enclosure=False)
pp(circles20)
circ.bubbles(circles20)
```
